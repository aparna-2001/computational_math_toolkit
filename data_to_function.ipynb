```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#converting a data to function
#ex 1
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])

#fitting into a straight line

m, c = np.polyfit(a,b,1)
print(m)
print(c)

def f(x):
  return m*x + c

print(f(5))
```

```python
#ex 2

x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([6,7,8,9,10])

def linear_regression(x,y):

    z = LinearRegression().fit(x,y)
    z_coeff = z.coef_
    z_intercept = z.intercept_
    print (f'coefficient : {z_coeff}')
    print (f'intercept : {z_intercept}')


linear_regression(x,y)

```
