```python
import csv

#creating data


data = [
    ["Name", "Math", "Physics", "Chemistry"],
    ["Alice", 85, 90, 78],
    ["Bob", 72, 88, 91],
    ["Carol", 95, 80, 85]
]

#opening a file to write

with open("student_data.csv", "w") as f:
  writer = csv.writer(f)
  writer.writerows(data)

#adding new row to the data - appending

with open("student_data.csv", "a") as f:
  writer = csv.writer(f)
  writer.writerow(["David", 16, 34, 56])


# Reopening the file in read mode to read its content,

with open("student_data.csv", "r") as f:
  rows = list(csv.reader(f))

rows.pop() # This modifies the 'rows' list in memory.

#modifying the change - writing it back

with open("student_data.csv", "w") as f:
  writer = csv.writer(f)
  writer.writerows(rows)

with open("student_data.csv","r") as f:
  print(f.read())
```

```python

# Product inventory data

products = [
    ["Product", "Price", "Stock", "Category"],
    ["Laptop", 899, 15, "Electronics"],
    ["Mouse", 25, 150, "Electronics"],
    ["Desk", 250, 8, "Furniture"],
    ["Chair", 180, 12, "Furniture"],
    ["Monitor", 320, 20, "Electronics"]
]

#data to product_data.csv
with open("product_data.csv", 'w') as file:
    writer = csv.writer(file)
    writer.writerows(products)

#appending a row
with open("product_data.csv", 'a') as file:
    writer = csv.writer(file)
    writer.writerow(["cot", 325, 432, "furniture"])

#deleting the third row
with open("product_data.csv", "r") as file:
  rows = list(csv.reader(file))
  rows.pop(3)

#writing the modifiled file
with open("product_data.csv", 'w') as file:
  writer = csv.writer(file)
  writer.writerows(rows)

#opening the final file
with open("product_data.csv", 'r') as file:
  print(file.read())
```
