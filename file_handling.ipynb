```python
import csv

#creating my own csv data

with open("my_data.csv","w") as f:
  writer = csv.writer(f)
  writer.writerows([['x','y','z'],[1,2,3],[4,5,6]])


#Reading the data I have created

with open("my_data.csv","r") as f:
  print(f.read())
```

```python
import csv

#creating data


data = [
    ["Name", "Math", "Physics", "Chemistry"],
    ["Alice", 85, 90, 78],
    ["Bob", 72, 88, 91],
    ["Carol", 95, 80, 85]
]

with open("student_data.csv", "w") as f:
  writer = csv.writer(f)
  writer.writerows(data)

with open("student_data.csv", "a") as f:
  writer = csv.writer(f)
  writer.writerow(["David", 16, 34, 56])

with open("student_data.csv","r") as f:
  print(f.read())

```
