```python
import pandas as pd
```

```python
with open("/content/Experiment Gene Expression.txt",'r') as file:
     lines = file.readlines()

lines[:10]
```

```python
#studying metadata starting with #

study_metadata = [line.strip() for line in lines if line.startswith('#')]
study_metadata
```

```python
#extracting the data

for i, line in enumerate(lines):
  if line.startswith("sampleID") or line.startswith('Gene'):
    print(i, line)
```

```python
#extracting dample metadata

sample_metadata = pd.read_csv("/content/Experiment Gene Expression.txt",
                              sep ='\t',
                              skiprows = range(0,5),
                              nrows = 4)

sample_metadata
```
