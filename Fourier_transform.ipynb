```python
#Fourier Transform tells us which brain frequency is dominant in the EEG signal.

#EEG data

import numpy as np
import matplotlib.pyplot as plt

fs = 100
t = np.arange(0, 1, 1/fs)

# EEG = alpha (10 Hz) + noise
eeg = np.sin(2*np.pi*10*t) + 0.3*np.random.randn(len(t))

print(eeg)
plt.plot(t, eeg)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python

#converting to frequency

fft_val = np.fft.fft(eeg)
print(fft_val)
```

```python
#frequency at which each fft value corresponds to

freq = np.fft.fftfreq(len(eeg), 1/fs)
print(freq)
```

```python
#finding the power

power = np.abs(fft_val)
print(power)
```

```python
import numpy as np
import pandas as pd
import sklearn as sk
```

```python
#opening the raw data of EEG and plotting it

raw = mne.io.read_raw_edf("S001R04.edf", preload =True)
raw.plot()
```

```python
#filtering the data
#filtering alpha and beta hertz

raw.filter(8,30)
```

```python
#extract events

events, event_id = mne.events_from_annotations(raw)
print(events)
#print(event_id)
```

```python
epochs = mne.Epochs(raw = raw,
           events = events,
           event_id = event_id,
           tmin = 0,
           tmax = 4,
           baseline = None,
           preload = True
)
```

```python
epochs_T0 = epochs['T0']
epochs_T2 = epochs['T2']

#print(epochs_T0)
#print(epochs_T2)

data_T0 = epochs_T0.get_data()
data_T2 = epochs_T2.get_data()

print(data_T0)
```

```python
fft_val_T0 = np.fft.fft(data_T0, axis=-1)
fft_val_T2 = np.fft.fft(data_T2, axis= -1)

print(fft_val_T0.shape)
print(fft_val_T2.shape)
```

```python

power_T0 = np.abs(fft_val_T0)
power_T2 = np.abs(fft_val_T2)

print(power_T0.shape)
print(power_T2.shape)
```

```python
 freqs = np.fft.fftfreq(data_T0.shape[-1], 1/fs)
 print(freqs)
```

```python
freqs = np.fft.fftfreq(data_T0.shape[-1], 1/fs)


mask_positive_freqs = freqs >= 0

power_T0 = power_T0[..., mask_positive_freqs]
power_T2 = power_T2[..., mask_positive_freqs]

freqs = freqs[mask_positive_freqs]
```

```python
alpha_T0 = power_T0[..., (freqs >= 8) & (freqs <= 13)].mean(axis=-1)
alpha_T2 = power_T2[..., (freqs >= 8) & (freqs <= 13)].mean(axis=-1)

print(alpha_T0.shape)
print(alpha_T2.shape)
```
