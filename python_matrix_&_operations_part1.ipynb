```python
#find the determinant of the below data
  #the matrix has been already generated


data_1 = {"sample":["s1","s2","s3"],
          "gene_1":[1,2,3],
          "gene_2":[4,5,6],
          "gene_3":[7,8,9]}

#converting this to table

import pandas as pd
table_1 = pd.DataFrame(data_1)
print(table_1)

#converting the table to matrix

matrix_4 = table_1[["gene_1","gene_2","gene_3"]].to_numpy()
print(matrix_4)

#FIND THE DETERMINANT OF matrix_4

import numpy as np
matrix_4 = np.array(matrix_4)
matrix_4_det = np.linalg.det(matrix_4)
print(f"Determinant of matrix_4: {matrix_4_det}")
```

**Since we found the determinant, what does the this determinat value actually means?**



*  If det is zero, which means rows/columns are linearly dependent
*  so only with matrix with rows/colums that are linearly independent has non zero determinat

*What would this dependent/independent mean in a data?*

Let's take our above code, (cell 32);
    - There the determinant is zero
        - which means the data may not be linearly independent
             - it may be linearly dependent
                  - they depend each other



**caution**: Only with determinat only we cannot say a data is correlated/dependent in real world problems

```python
#we can use 'linalg' module to find many features of a matrix(or data)
   #inverse
   #Eigen values and Eigen vectors
   #Rank


import numpy as np

#lets find inverse of a matrix
    #key point to remember: inverse only exist for matrix with determinant having non zero value

#lets find the inverse
data_2 = [
    [2, 1, 1],
    [1, 3, 2],
    [1, 0, 1]
]

data_2 = np.array(data_2)

#determinant of the matrix
data_2_det = np.linalg.det(data_2)
print(f"Determinant of data_2 matrix: {data_2_det}")

data_2_inv = np.linalg.inv(data_2)
print(f"Inverse of the matrix data_2: {data_2_inv}")
```

```python
#lets try to find the inverse of a matrix of determinant 0

data_1 = {"sample":["s1","s2","s3"],
          "gene_1":[1,2,3],
          "gene_2":[4,5,6],
          "gene_3":[7,8,9]}

#converting this to table

import pandas as pd
table_1 = pd.DataFrame(data_1)
print(table_1)

#converting the table to matrix

matrix_4 = table_1[["gene_1","gene_2","gene_3"]].to_numpy()
print(matrix_4)

#FIND THE DETERMINANT OF matrix_4

import numpy as np
matrix_4 = np.array(matrix_4)
matrix_4_det = np.linalg.det(matrix_4)
print(f"Determinant of matrix_4: {matrix_4_det}")

#using try except to catch the expected error
try:
    matrix_4_inv = np.linalg.inv(matrix_4)
    print(matrix_4_inv)
except:
    print("singular matrix")
```

```python
#Finding Eigen values and Eigen vector
  #Eigen values represent the variance along corresponding Eigen vector
    #larger the value: more the variance along that vector
    #lesser or negative value: less the variance along that vector

import pandas as pd
import numpy as np



matrix_ex1 = [
    [4, 2, 1, 3, 5],
    [0, 1, 2, 4,5],
    [2, 0, 3, 1, 2],
    [1, 2, 0, 2, 3],
    [3, 1, 2, 0, 4]
]

matrix_ex1 = np.array(matrix_ex1)
matrix_ex1_eigenvalue, matrix_ex1_eigenvector = np.linalg.eig(matrix_ex1)
print(f"The Eigen values are: {matrix_ex1_eigenvalue}\n")
print(f"The Eigen vectors are: {matrix_ex1_eigenvector}")

#usually in data analysis, Eigen values/vectors are calculated for covariance matrix. Not raw data


```

```python
#practicing Eigen values and Eigen vectors

data_behavior = {
    "participant": ["P1", "P2", "P3", "P4"],
    "reaction_time_ms": [350, 420, 300, 390],
    "accuracy_percent": [95, 88, 100, 92],
    "trial_count": [10, 12, 9, 11],
    "response_time":[6,8,9,11]
}

import pandas as pd
import numpy as np

table_3 = pd.DataFrame(data_behavior)
print(table_3)

matrix_3 = np.array(table_3[["reaction_time_ms","accuracy_percent","trial_count","response_time"]])
print(matrix_3)

matrix_3_eigenvalue, matrix_3_eigenvector = np.linalg.eig(matrix_3)
print(f"Eigen value of the matrix_3: {matrix_3_eigenvalue}")
print(f"Eigen vector of the matrix_3:{matrix_3_eigenvector}")
```

```python
#Eigen values and Eigen vectors

#To refresh our memory, we will start with this data
    #convert it to table
       #then to matrix
          #then we will find the Eigen values and Eigen vector

data_behavior = {
    "participant": ["P1", "P2", "P3", "P4"],
    "reaction_time_ms": [350, 420, 300, 390],
    "accuracy_percent": [95, 88, 100, 92],
    "trial_count": [10, 12, 9, 11]
}

import pandas as pd
import numpy as np

table_3 = pd.DataFrame(data_behavior)
print(table_3)

matrix_6 = np.array(table_3[["reaction_time_ms","accuracy_percent","trial_count"]])
print(matrix_6)

try:
    matrix_6_det = np.linalg.det(matrix_6)
    print(f"Determinant of matrix_6: {matrix_6_det}")
except:
    print("This is not a square matrix")

#since this is not a square matrix
  #no determinant
    #since there is no determinant
       #then there is no inverse
         #there is no eigen values or eigen vector either
```

```python
#now let's try to find Rank

#In the above matrix, we see that we cannot calculate det,inv,eigen values or eigen vectors. since it was not a square matrix

#let's try to find it's rank

data_behavior = {
    "participant": ["P1", "P2", "P3", "P4"],
    "reaction_time_ms": [350, 420, 300, 390],
    "accuracy_percent": [95, 88, 100, 92],
    "trial_count": [10, 12, 9, 11]
}

import pandas as pd
import numpy as np

table_3 = pd.DataFrame(data_behavior)
print(table_3)

matrix_6 = np.array(table_3[["reaction_time_ms","accuracy_percent","trial_count"]])
print(matrix_6)

try:
    matrix_6_det = np.linalg.det(matrix_6)
    print(f"Determinant of matrix_6: {matrix_6_det}")
except:
    print("This is not a square matrix")

#finding rank

matrix_6_rank = np.linalg.matrix_rank(matrix_6)
print(f"Rank of the matrix: {matrix_6_rank}")
```

**What does it mean by Rank?**

- In the above code, Rank 3 means there are 3  columns are linearly independent, that we cannot transform one variable to another,

therefore each variable holds independent information and essential to data

```python
#Now what if we take another data and find it's rank

data_example = {
    "x": [1, 2, 3, 4],
    "y": [2, 4, 6, 8],
    "z": [5, 6, 7, 8]
}

import numpy as np
import pandas as pd

table_7 =pd.DataFrame(data_example)
print(table_7)

matrix_7 = np.array(table_7[["x","y","z"]])
print(matrix_7)

try:
  matrix_7_det = np.linalg.det(matrix_7)
  print(f"Determinat of matrix_7: {matrix_7_det}")
except:
  print("This is not a square matrix")

matrix_7_rank = np.linalg.matrix_rank(matrix_7)
print(f"Rank of the matrix: {matrix_7_rank}")

#Here rank 2 means, only two columns are linearly independent.
```
