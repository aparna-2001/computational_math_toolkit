```python
#linear regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(1)
n = 100

hours_studied = np.random.uniform(0, 10, n)
test_scores = 6.2 * hours_studied + 48 + np.random.normal(0, 5, n)

# Keep scores in realistic range
test_scores = np.clip(test_scores, 0, 100)


data_array = np.array([hours_studied,test_scores])

slope,intercept = np.polyfit(hours_studied,test_scores,1)
#print(f"slope:{slope}")
#print(f"intercept:{intercept}")

corr_coeff = np.corrcoef(hours_studied,test_scores)[0,1]**2
#print(corr_coeff)

predicted_test_scores = slope* np.array(hours_studied)+intercept
#print(predicted_test_scores)

errors = test_scores - predicted_test_scores
mean_squared_error = np.mean(errors**2)
sqrt_mse = np.sqrt(mean_squared_error)
print(sqrt_mse)
#print(mean_squared_error)

z = slope*np.array(hours_studied)+intercept


def draw_scatter_plot(x,y):
  plt.title("scatter plot for study hours and test scores")
  plt.xlabel("Hours_studied")
  plt.ylabel("Test_scores")
  plt.scatter(x,y)
  #plt.plot(x,z, color='red')


  x_line = np.linspace(min(x), max(x), 100)
  y_line = slope * x_line + intercept
  plt.plot(x_line, y_line, color="red")
  plt.show()
draw_scatter_plot(hours_studied,test_scores)
```
