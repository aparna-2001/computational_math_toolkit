```python
#converting the linear equation to matrix



#we are using a new package called sympy
  #for the conversion of linear equations to matrix


from sympy import symbols, Eq, linear_eq_to_matrix

x, y = symbols('x y')

eqn_1 = Eq(2*x + 3*y, 8)
eqn_2 = Eq(5*x + 4*y, 13)

eqn_1_matrix = linear_eq_to_matrix([eqn_1],[x,y])
eqn_2_matrix = linear_eq_to_matrix([eqn_2], [x,y])

print(f"matrix of eqn_1: {eqn_1_matrix}")
print(f"matrix of eqn 2: {eqn_2_matrix}")
```

```python
# To solve linear equation it is not much use writing as separate matrices

#let's merge it to one matrix

from sympy import Eq, symbols, linear_eq_to_matrix

x, y = symbols('x y')

eqn_1 = Eq(2*x + 3*y, 8)
eqn_2 = Eq(5*x + 4*y, 13)


#combining everything here

eqn_matrix_with_coeff, eqn_matrix_without_coeff = linear_eq_to_matrix([eqn_1, eqn_2],[x,y])

print(eqn_matrix_with_coeff)
print(eqn_matrix_without_coeff)

```
