```python
#dummy eeg data

import numpy as np
import pandas as pd

np.random.seed(0)

n_samples = 5000

signal_data_set = pd.DataFrame({
    'FP1': np.random.normal(0, 50, n_samples),
    'FP2': np.random.normal(5, 60, n_samples),
    'FC5': np.random.normal(-10, 40, n_samples),
    'FC6': np.random.normal(0, 55, n_samples),
    'FZ': np.random.normal(3, 45, n_samples),
    'CZ': np.random.normal(-5, 35, n_samples)
})

signal_data_set.head()
```

```python
import matplotlib.pyplot as plt

def plot_hist(data_series, ax):
    ax.hist(data_series)
    ax.set_title(f'Histogram of {data_series.name}')
    ax.set_xlabel('amplitudes(in microvolts)')
    ax.set_ylabel('frequency')

headers = ['FP1', 'FP2', 'FC5', 'FC6', 'FZ', 'CZ']

# Create subplots for each channel
fig, axes = plt.subplots(nrows=len(headers), ncols=1, figsize=(10, 4 * len(headers)))
fig.tight_layout(pad=3.0) # Adjust layout to prevent overlapping titles/labels

for i, header in enumerate(headers):
    plot_hist(signal_data_set[header], axes[i]) # Pass individual Series
plt.show()
```

```python
import numpy as np
import pandas as pd

np.random.seed(0)

n_students = 500

student_data = pd.DataFrame({
    'Math_Score': np.random.normal(70, 10, n_students),
    'Physics_Score': np.random.normal(65, 12, n_students),
    'Chemistry_Score': np.random.normal(68, 11, n_students),
    'Study_Hours': np.random.normal(5, 1.5, n_students),
    'Sleep_Hours': np.random.normal(7, 1, n_students)
})

student_data.head()
```

```python
def plt_histo(data, ax):
  ax.hist(data)
  ax.set_title(f"Histogram of {data.name}")
  ax.set_xlabel('marks')
  ax.set_ylabel('frequency')


data_headers = ['Math_Score',	'Physics_Score'	,'Chemistry_Score',	'Study_Hours', 'Sleep_Hours']

fig, axes = plt.subplots(nrows=len(data_headers), ncols=1, figsize=(10, 4 * len(data_headers)))
fig.tight_layout(pad=3.0)

for i, header in enumerate(data_headers):
  plt_histo(student_data[header], ax=axes[i])
plt.show()
```
