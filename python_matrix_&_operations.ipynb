```python
#calculating the determinant of a matrix

import numpy as np

#here we are using 'linalg'module in numpy

matrix_2 = [[1,2,3],[4,5,6],[7,8,9]]
matrix_2 = np.array(matrix_2)

det = np.linalg.det(matrix_2)
print(f"Determinant of matrix_2: {det}")
```

```python
#find the determinant of the below data
  #the matrix has been already generated


data_1 = {"sample":["s1","s2","s3"],
          "gene_1":[1,2,3],
          "gene_2":[4,5,6],
          "gene_3":[7,8,9]}

#converting this to table

import pandas as pd
table_1 = pd.DataFrame(data_1)
print(table_1)

#converting the table to matrix

matrix_4 = table_1[["gene_1","gene_2","gene_3"]].to_numpy()
print(matrix_4)

#FIND THE DETERMINANT OF matrix_4

import numpy as np
matrix_4 = np.array(matrix_4)
matrix_4_det = np.linalg.det(matrix_4)
print(f"Determinant of matrix_4: {matrix_4_det}")
```

**Since we found the determinant, what does the this determinat value actually means?**



*  If det is zero, which means rows/columns are linearly dependent
*  so only with matrix with rows/colums that are linearly independent has non zero determinat

*What would this dependent/independent mean in a data?*

Let's take our above code, (cell 32);
    - There the determinant is zero
        - which means the data may not be linearly independent
             - it may be linearly dependent
                  - they depend each other



**caution**: Only with determinat only we cannot say a data is correlated/dependent in real world problems
