```python
#input of data

import numpy as np

matrix_1 = input(np.array([],[]))
```

```python
# finding the determinant

import numpy as np

def det_matrix():
    try:
      matrix_1 = np.linalg.det(matrix_1)
      print(matrix_1)
    except:
      print("This is a non square matrix")
```

```python
#finding the inverse

def inverse_matrix():
  try:
    matrix_inv = np.linalg.inv(matrix_1)
    print(matrix_inv)
  except:
    print('The matrix is singular or The matrix is not square matrix')
```

```python
#finding the eigen values and eigen vectors

def eigen_matrix():
  try:
    Eigen_values,Eigen_vectors = np.linalg.eig(matrix_1)
    print(Eigen_values)
    print(Eigen_vectors)
  except:
    print('The matrix is not a square matrix')
```

```python
#finding the rank

def rank_matrix():
    rank = np.linalg.matrix_rank(matrix_1)
    print(rank)
```
